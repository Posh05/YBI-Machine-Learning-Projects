```python
y = df['Item_Outlet_Sales']
y.shape
```

```python
x = df.drop(['Item_Identifier','Item_Outlet_Sales'],axis = 1)
x.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x_standard = x[['Item_Weight','Item_Visibility','Item_MRP','Outlet_Establishment_Year']]
```

```python
x_standard = sc.fit_transform(x_standard)
```

```python
x_standard
```

```python
x[['Item_Weight','Item_Visibility','Item_MRP','Outlet_Establishment_Year']] = pd.DataFrame(x_standard,columns = [['Item_Weight','Item_Visibility','Item_MRP','Outlet_Establishment_Year']])
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 192529)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape 
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(random_state = 192529)
```

```python
rfr.fit(x_train,y_train)
```

```python
predictions = rfr.predict(x_test)
predictions
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,predictions)
```

```python
plt.figure(figsize = (10,8))
plt.scatter(y_test,predictions)
plt.xlabel('Actual Sales')
plt.ylabel('Predicted Sales')
```
